## Exercise 10 "What Was That?"

For this exercise the explanation of exercise 9 is extended with the use of the character \ to put a new character in the same string. The / character helps Python avoid getting confused with the use of double or single quotes in a string.



```python
me = "\t I'm very tired."
you= "you are \very handsome."
metoo = "I'm \\ a \\ great person."

your = """
I'll do a list:
\t* you
\t* we
\t* they \n\t* she
 """

print(me)
print(you)
print(metoo)
print(your)
```
